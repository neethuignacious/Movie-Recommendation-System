```python
import numpy as np
import pandas as pd
```

```python
credits_df = pd.read_csv('credits.csv')
```

```python
credits_df.head()
```

```python
meta_df = pd.read_csv('movies_metadata.csv');
```

```python
meta_df.tail()
```

```python
meta_df['release_date'] = pd.to_datetime(meta_df['release_date'],errors='coerce')
```

```python
meta_df['year'] = meta_df['release_date'].dt.year
```

```python
movie_df = meta_df.loc[meta_df.year <= 2017,['genres','id','title','year']]
```

```python
movie_df.head()
```

```python
movie_df.dtypes
```

```python
movie_df['id'] = movie_df['id'].astype(int)
```

```python
movie_df.dtypes
```

```python
credits_df.dtypes
```

```python
new_df = pd.merge(movie_df, credits_df, on='id')
```

```python
new_df.head()
```

```python
import ast
new_df['genres'] = new_df['genres'].map(lambda x: ast.literal_eval(x))
new_df['cast'] = new_df['cast'].map(lambda x: ast.literal_eval(x))
new_df['crew'] = new_df['crew'].map(lambda x: ast.literal_eval(x))
```

```python
def genres_list(x):
    gen_lt = []
    c = " "
    for i in x:
        if i.get('name') == "Science Fiction":
            gen_lt.append('Sci-Fi')
        else:
            gen_lt.append(i.get('name'))
    if gen_lt == []:
        return np.NaN
    else:
        return (c.join(gen_lt))
```

```python
new_df['genres'] = new_df['genres'].map(lambda x: genres_list(x))
```

```python
new_df.head()
```

```python
def actor1(x):
    actor = []
    for i in x:
        actor.append(i.get('name'))
    if actor == []:
        return np.NaN
    else:
        return (actor[0])
    
new_df['actor1'] = new_df['cast'].map(lambda x: actor1(x))
```

```python
def actor2(x):
    actor = []
    for i in x:
        actor.append(i.get('name'))
    if actor == [] or len(actor)<=1:
        return np.NaN
    else:
        return (actor[1])
new_df['actor2'] = new_df['cast'].map(lambda x: actor2(x))
```

```python
def actor3(x):
    actor = []
    for i in x:
        actor.append(i.get('name'))
    if actor == [] or len(actor)<=2:
        return np.NaN
    else:
        return (actor[2])

new_df['actor3'] = new_df['cast'].map(lambda x: actor3(x))
```

```python
new_df.head()
```

```python
def director(x):
    lt = []
    c = " "
    for i in x:
        if i.get('job') == 'Director':
            lt.append(i.get('name'))
    if lt == []:
        return np.NaN
    else:
        return (c.join(lt))
    
new_df['director'] = new_df['crew'].map(lambda x: director(x))
```

```python
new_df.head()
```

```python
movie = new_df.loc[:,['director','actor1','actor2','actor3','genres','title']]
```

```python
movie['combined'] = movie['actor1'] + ' ' + movie['actor2'] + ' '+ movie['actor3'] + ' '+ movie['director'] +' ' + movie['genres']
```

```python
movie = movie.dropna(how='any')
```

```python
movie.drop_duplicates(subset ="title", keep = 'last', inplace = True)
```

```python
movie.head()
```

```python
movie.to_csv('movie.csv',index=False)
```
